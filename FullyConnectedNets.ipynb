As a sanity check, run the following to check the initial loss and to gradient check the network both with and without regularization. This is a good way to see if the initial losses seem reasonable.

For gradient checking, you should expect to see errors around 1e-7 or less.

````python
np.random.seed(231)
N, D, H1, H2, C = 2, 15, 20, 30, 10
X = np.random.randn(N, D)
y = np.random.randint(C, size=(N,))

for reg in [0, 3.14]:
    print("Running check with reg = ", reg)
    model = FullyConnectedNet(
        [H1, H2],
        input_dim=D,
        num_classes=C,
        reg=reg,
        weight_scale=5e-2,
        dtype=np.float64
    )

    loss, grads = model.loss(X, y)
    print("Initial loss: ", loss)

    # Most of the errors should be on the order of e-7 or smaller.
    # NOTE: It is fine however to see an error for W2 on the order of e-5
    # for the check when reg = 0.0
    for name in sorted(grads):
        f = lambda _: model.loss(X, y)[0]
        grad_num = eval_numerical_gradient(f, model.params[name], verbose=False, h=1e-5)
        print(f"{name} relative error: {rel_error(grad_num, grads[name])}")
````

## Answer:
When comparing the training difficulty of these two networks, a five-layer network is more sensitive to weight initialization. This is mainly due to the increased number of layers in the network. In deeper networks, if the scale of initialization is set too small, it can lead to vanishing gradients (where the product of gradients becomes very small), whereas if it is set too large, it may cause exploding gradients (where the gradient's product becomes overly large). This situation is more common in deeper networks than in shallower ones.

## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?

## Answer
AdaGrad updates become very small because the squared gradients accumulate over time. As training progresses, the cache value, which stores the sum of squared gradients, grows larger since it keeps adding positive values. Consequently, dividing the gradient (dw) by the square root of the cache results in very small updates, slowing down the learning process.

Adam, however, mitigates this issue by using an exponential weighted average of the squared gradients, similar to RMSprop. This approach gives more weight to recent gradients and less weight to older ones, preventing the cache from growing too large. Additionally, Adam combines this with the momentum term, which helps maintain larger updates controlled by the exponential weighted average of the squared gradients.
